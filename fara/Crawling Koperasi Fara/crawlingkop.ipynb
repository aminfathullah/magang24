Penarikan data diatas bertujuan untuk menampilkan ID Koperasi secara menyeluruh (unlock), kemudian dari ID Koperasi lengkap diatas ditarik untuk menghasilkan data di dalam detail tiap koperasi seperti di bawah ini.

```python
import re
import requests
import threading
import pandas as pd

def fetch_data(page, cooperative_data):
    url = 'https://data.diskopukm.jatimprov.go.id/satu_data/webnya/ajax_listOds'
    payload = {
        'draw': 1,
        'columns[0][data]': '0',
        'columns[0][name]': '',
        'columns[0][searchable]': 'true',
        'columns[0][orderable]': 'true',
        'columns[0][search][value]': '',
        'columns[0][search][regex]': 'false',
        'order[0][column]': '0',
        'order[0][dir]': 'asc',  
        'start': page,
        'length': 10,
        'search[value]': '',
        'search[regex]': 'false',
    }
    response = requests.post(url, data=payload)
    if response.status_code == 200:
        data = response.json()
        cooperative_data.extend(data['data'])
    else:
        print("Failed to fetch data")

def extract_id_kop(text):
    match = re.search(r'(\d+)', text)
    if match:
        return match.group(1)
    else:
        return None

def fetch_data_detail(base_url, coop_id, cooperative_data):
    url = base_url + coop_id
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        cooperative_data.append(data)
    else:
        cooperative_data.append(None)

def main():
    cooperative_data = []
    threads = []
    num_data = 20
    num_requests = num_data // 10 + (num_data % 10 > 0)
    for i in range(num_requests):
        page = i * 10
        thread = threading.Thread(target=fetch_data, args=(page, cooperative_data))
        thread.start()
        threads.append(thread)

    for thread in threads:
        thread.join()

    column_titles = ['NO', 'ID KOP', 'KOPERASI', 'ALAMAT', 'NO BADAN HUKUM', 'TANGGAL BADAN HUKUM', 'DESA', 'KELURAHAN', 'AKSI']

    df = pd.DataFrame(cooperative_data, columns=column_titles)
    
    df['ID KOP'] = df['AKSI'].apply(extract_id_kop)
    
    pd.set_option('display.max_colwidth', None)

    print("Data  Koperasi:")
    print(df.to_string(index=False))

    base_url = 'https://data.diskopukm.jatimprov.go.id/satu_data/webnya/detailOds/'
    id_kops = df['ID KOP'].tolist()

    cooperative_data = []
    threads = []
    for id_kop in id_kops:
        thread = threading.Thread(target=fetch_data_detail, args=(base_url, id_kop, cooperative_data))
        thread.start()
        threads.append(thread)

    for thread in threads:
        thread.join()

    df_detail = pd.DataFrame(cooperative_data)

    print("\nData Detail Koperasi:")
    print(df_detail.to_string(index=False))

if __name__ == "__main__":
    main()
```
